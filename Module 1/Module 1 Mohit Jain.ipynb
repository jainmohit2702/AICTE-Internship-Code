<h6> Question 1.5 : </h6>

```python
#1.5 - 
#Calculate Avg Price, Profit and Loss Percent for last N days
```

```python
def avg_price(no_of_days):
    avg_n_days = (df1['Average Price'].tail(no_of_days).sum())/no_of_days
    return avg_n_days
n = int(input("Input no. of days = "))
avg_price(n)
```

```python
def profit_loss_percent(no_of_days):
    df5 = df1['Close Price'].tail(no_of_days)
    final_value = np.array(df5.tail(1))
    initial_value = np.array(df5.head(1))
    p_l = float(((final_value - initial_value)/initial_value)*100)
    p_l = round(p_l,3)
    if(p_l<0):
        print("Loss Percentage = ",abs(p_l),"%")
    elif(p_l==0):
        print("Neither Profit Nor Loss = ",p_l,"%")
    else:
        print("Profit Percentage = ",p_l,"%")
        

N = int(input("Input number of days = "))        
profit_loss_percent(N)
```

```python
# For 1 WEEK :
print("Average Price = ",avg_price(7))
print(profit_loss_percent(7))
```

```python
# For 2 WEEKS :
print("Average Price = ",avg_price(14))
print(profit_loss_percent(14))
```

```python
# For 1 Month (Considering 1 month contains 30 days) :
print("Average Price = ",avg_price(30))
print(profit_loss_percent(30))
```

```python
#1.6 - 
# Find out percent change between 2 consecutive close prices and add a column
# Use pct_change() function to do the same.
```

```python
perc=df1[['Close Price']].pct_change()
perc=perc*100           # we multiply by 100 to get the change in percentage
```

```python
# The first value is NAN because the pct_change function compares two values, 
# thats why for row 0 there is no comparable data.
# So we relace the NAN value by 0 we use fillna() function of pandas and replace by 0
df1['Day_Perc_Change']=perc.fillna(0)
```

```python
df=np.array(df1['Day_Perc_Change'])
Trend=[]
for i in range(0,494):
    if df[i]>=-0.5 and df[i]<0.5:
        Trend.append('Slight or No Change')
    elif df[i]>=0.5 and df[i]<1.0:
        Trend.append('Slight Positive')
    elif df[i]<=-0.5 and df[i]>-1.0:
        Trend.append('Slight Negative')
    elif df[i]>=1.0 and df[i]<3.0:
        Trend.append('Positive')
    elif df[i]<=-1.0 and df[i]>-3.0:
        Trend.append('Negative')
    elif df[i]>=3.0 and df[i]<=7.0:
        Trend.append('Among top gainers')
    elif df[i]<=-3.0 and df[i]>-7.0:
        Trend.append('Among top losers')
    elif df[i] >7.000:
        Trend.append('Bull run')
    elif df[i]<=-7.000:
        Trend.append('Bear drop')
```

```python
df7 = pd.DataFrame(Trend,columns=['Trend'])
df1 = pd.concat([df1,df7],axis=1)
df1
```
